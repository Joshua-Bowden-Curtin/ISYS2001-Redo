````python
# AI-assisted data loading section
# (Copy the code AI provides here)

# Add your own comments about what you learned

# ============================================================
# BLOCK 1: Data Loading & Cleaning
# Smart Finance Assistant – ISYS2001 Final Project
# ============================================================

import pandas as pd
import os

# ------------------------------------------------------------
# STEP 1: Load the CSV file
# ------------------------------------------------------------

# Define the path to the transactions CSV file.
# Update this path if your file is stored elsewhere (e.g., mounted in Colab).
CSV_PATH = "transactions.csv"

def load_transactions(filepath):
    """
    Load transaction data from a CSV file into a pandas DataFrame.

    Parameters:
        filepath (str): Path to the CSV file.

    Returns:
        pd.DataFrame: Raw transaction data, or None if loading fails.
    """
    # Check the file exists before attempting to load it
    if not os.path.exists(filepath):
        print(f"[ERROR] File not found: '{filepath}'")
        print("        Please check the file path and try again.")
        return None

    try:
        df = pd.read_csv(filepath)
        print(f"[OK] Loaded '{filepath}' — {len(df)} rows, {len(df.columns)} columns.")
        return df
    except Exception as e:
        print(f"[ERROR] Could not read CSV: {e}")
        return None


# ------------------------------------------------------------
# STEP 2: Clean & Validate the Data
# ------------------------------------------------------------

def clean_transactions(df):
    """
    Clean and validate a raw transaction DataFrame.

    Cleaning steps:
        1. Strip whitespace from column names and string values.
        2. Remove the '$' sign (and any commas) from the Amount column,
           then convert it to a numeric float.
        3. Parse the Date column into proper datetime objects.
        4. Drop rows where Amount or Date are missing/unparseable (critical fields).
        5. Fill missing Category or Description with sensible defaults.
        6. Remove exact duplicate rows.

    Parameters:
        df (pd.DataFrame): Raw DataFrame from load_transactions().

    Returns:
        pd.DataFrame: Cleaned DataFrame ready for analysis.
    """
    if df is None or df.empty:
        print("[WARNING] No data to clean.")
        return df

    # Work on a copy so the original raw data is preserved for debugging
    df = df.copy()

    # -- 2a. Normalise column names ---------------------------------
    # Strip surrounding whitespace and standardise capitalisation
    df.columns = df.columns.str.strip()

    # Validate that the required columns are present
    required_columns = {"Date", "Amount", "Category", "Description"}
    missing_cols = required_columns - set(df.columns)
    if missing_cols:
        print(f"[ERROR] CSV is missing required columns: {missing_cols}")
        return None

    # -- 2b. Strip whitespace from all string (object) columns ------
    string_cols = df.select_dtypes(include="object").columns
    for col in string_cols:
        df[col] = df[col].str.strip()

    # -- 2c. Clean the Amount column --------------------------------
    # Raw values may look like "$1,250.00" or "- $45.99"
    # Remove '$', commas, and any stray spaces before converting
    df["Amount"] = (
        df["Amount"]
        .astype(str)                     # ensure we can call string methods
        .str.replace(r"[$,\s]", "", regex=True)  # remove $, commas, spaces
    )
    # Convert to numeric; non-numeric entries become NaN
    df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

    # -- 2d. Parse the Date column ----------------------------------
    # errors="coerce" turns unparseable dates into NaT (Not a Time)
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

    # -- 2e. Drop rows with missing critical fields -----------------
    before = len(df)
    df = df.dropna(subset=["Date", "Amount"])
    dropped = before - len(df)
    if dropped > 0:
        print(f"[WARNING] Dropped {dropped} row(s) with invalid Date or Amount.")

    # -- 2f. Fill optional missing fields ---------------------------
    df["Category"]    = df["Category"].fillna("Uncategorised")
    df["Description"] = df["Description"].fillna("No description")

    # -- 2g. Remove exact duplicate rows ----------------------------
    dupes = df.duplicated().sum()
    if dupes > 0:
        df = df.drop_duplicates()
        print(f"[WARNING] Removed {dupes} duplicate row(s).")

    # Reset the index so it runs 0, 1, 2, … after any dropped rows
    df = df.reset_index(drop=True)

    print(f"[OK] Cleaning complete — {len(df)} valid rows remaining.")
    return df


# ------------------------------------------------------------
# STEP 3: Show Basic Dataset Information
# ------------------------------------------------------------

def summarise_dataset(df):
    """
    Print a human-readable summary of the cleaned transaction DataFrame.

    Displays:
        - Row and column counts
        - Date range covered
        - Total, average, min, and max transaction amounts
        - Number of unique spending categories
        - Count of any remaining null values per column

    Parameters:
        df (pd.DataFrame): Cleaned transaction DataFrame.
    """
    if df is None or df.empty:
        print("[WARNING] No data available to summarise.")
        return

    print("\n" + "=" * 50)
    print("  DATASET SUMMARY")
    print("=" * 50)

    print(f"  Rows         : {len(df)}")
    print(f"  Columns      : {list(df.columns)}")
    print(f"  Date range   : {df['Date'].min().date()}  →  {df['Date'].max().date()}")
    print(f"  Total spend  : ${df['Amount'].sum():,.2f}")
    print(f"  Average txn  : ${df['Amount'].mean():,.2f}")
    print(f"  Largest txn  : ${df['Amount'].max():,.2f}")
    print(f"  Smallest txn : ${df['Amount'].min():,.2f}")
    print(f"  Categories   : {df['Category'].nunique()} unique")
    print(f"    → {sorted(df['Category'].unique())}")

    # Highlight any remaining null values (should be none after cleaning)
    null_counts = df.isnull().sum()
    if null_counts.any():
        print(f"\n  [WARNING] Remaining nulls:\n{null_counts[null_counts > 0]}")
    else:
        print("\n  [OK] No null values remaining.")

    print("=" * 50 + "\n")

    # Show the first few rows so the student can visually confirm the data
    print("  First 5 rows:")
    print(df.head().to_string(index=False))
    print()


# ------------------------------------------------------------
# MAIN – Run all three steps
# ------------------------------------------------------------

raw_df   = load_transactions(CSV_PATH)
clean_df = clean_transactions(raw_df)
summarise_dataset(clean_df)
````

````python
# AI-generated analysis functions
# (Your AI collaboration code goes here)

# ============================================================
# BLOCK 2: Spending Analysis Functions
# Smart Finance Assistant – ISYS2001 Final Project
# ============================================================
# Business Purpose:
#   These functions answer the core personal finance questions:
#   "Where is my money going?" and "Is anything unusual happening?"
#   Each function is independent so the main block can call any
#   combination depending on what the user asks for.
# ============================================================

import pandas as pd


# ------------------------------------------------------------
# HELPER: Validate that the DataFrame is usable
# ------------------------------------------------------------

def _validate_df(df, required_cols):
    """
    Internal helper — checks that df is a non-empty DataFrame
    containing the columns each analysis function needs.

    Parameters:
        df              : the object to validate
        required_cols   : list of column names that must be present

    Returns:
        True if valid, False otherwise (and prints a clear message).
    """
    if df is None or not isinstance(df, pd.DataFrame) or df.empty:
        print("[ERROR] No transaction data available. "
              "Please run the loading & cleaning block first.")
        return False

    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        print(f"[ERROR] DataFrame is missing required columns: {missing}")
        return False

    return True

````

````python
# ------------------------------------------------------------
# FUNCTION 1: Total Spending by Category
# ------------------------------------------------------------

def spending_by_category(df):
    """
    Calculate the total amount spent in each spending category.

    Business Purpose:
        Tells the user exactly how much they have spent in areas like
        Groceries, Rent, Entertainment, etc. — the foundation of any
        budget review.

    How it works:
        Groups all transactions by the Category column and sums the
        Amount column within each group. Results are sorted largest
        first so the biggest cost centres are immediately visible.

    Parameters:
        df (pd.DataFrame): Cleaned transaction DataFrame from Block 1.

    Returns:
        pd.DataFrame: Two-column table — Category and Total_Amount —
                      sorted descending by Total_Amount.
                      Returns None if the input is invalid.
    """
    if not _validate_df(df, ["Category", "Amount"]):
        return None

    # Group by Category and sum the Amount column
    category_totals = (
        df.groupby("Category", as_index=False)["Amount"]
        .sum()
        .rename(columns={"Amount": "Total_Amount"})
        .sort_values("Total_Amount", ascending=False)
        .reset_index(drop=True)
    )

    # Round to 2 decimal places for clean dollar display
    category_totals["Total_Amount"] = category_totals["Total_Amount"].round(2)

    print("\n📊 TOTAL SPENDING BY CATEGORY")
    print("-" * 35)
    for _, row in category_totals.iterrows():
        print(f"  {row['Category']:<20} ${row['Total_Amount']:>10,.2f}")
    print("-" * 35)
    print(f"  {'TOTAL':<20} ${category_totals['Total_Amount'].sum():>10,.2f}\n")

    return category_totals
````

````python
# ------------------------------------------------------------
# FUNCTION 2: Top 5 Spending Categories
# ------------------------------------------------------------

def top_spending_categories(df, n=5):
    """
    Identify the top N categories where the most money was spent.

    Business Purpose:
        Highlights the biggest drains on a user's budget at a glance.
        For example, knowing that 60 % of spending is on Rent and Food
        helps prioritise where to cut back.

    How it works:
        Reuses spending_by_category() and takes the first n rows,
        then adds a percentage column so each category's share of
        total spending is clear.

    Parameters:
        df (pd.DataFrame): Cleaned transaction DataFrame from Block 1.
        n  (int)         : Number of top categories to return (default 5).

    Returns:
        pd.DataFrame: Top-N categories with Total_Amount and Percentage
                      columns, sorted descending.
                      Returns None if the input is invalid.
    """
    # Re-use Function 1 rather than repeating the groupby logic
    category_totals = spending_by_category(df)

    if category_totals is None or category_totals.empty:
        return None

    # Clamp n so we never ask for more rows than exist
    n = min(n, len(category_totals))
    top_n = category_totals.head(n).copy()

    # Add a percentage column — useful for pie charts later
    grand_total = category_totals["Total_Amount"].sum()
    if grand_total == 0:
        top_n["Percentage"] = 0.0
    else:
        top_n["Percentage"] = ((top_n["Total_Amount"] / grand_total) * 100).round(1)

    print(f"\n🏆 TOP {n} SPENDING CATEGORIES")
    print("-" * 45)
    for rank, (_, row) in enumerate(top_n.iterrows(), start=1):
        bar = "█" * int(row["Percentage"] / 5)   # simple ASCII bar chart
        print(f"  {rank}. {row['Category']:<18} ${row['Total_Amount']:>9,.2f}"
              f"  ({row['Percentage']}%) {bar}")
    print()

    return top_n
````

````python
# ------------------------------------------------------------
# FUNCTION 3: Average Spending Per Transaction
# ------------------------------------------------------------

def average_spending_per_transaction(df, by_category=False):
    """
    Calculate the average (mean) transaction amount, either overall
    or broken down by category.

    Business Purpose:
        The average transaction size helps spot lifestyle creep —
        if the average grocery trip costs $180 instead of $90,
        something has changed. Per-category averages also help
        identify which spending areas have the most variability.

    How it works:
        Uses pandas .mean() on the Amount column. When by_category
        is True, it applies a groupby first, then sorts by average
        descending so the most expensive transaction types are visible.

    Parameters:
        df          (pd.DataFrame): Cleaned transaction DataFrame.
        by_category (bool)        : If True, break averages down
                                    by Category (default False).

    Returns:
        float or pd.DataFrame:
            - Overall average (float) when by_category=False.
            - Per-category averages (DataFrame) when by_category=True.
            Returns None if the input is invalid.
    """
    if not _validate_df(df, ["Amount", "Category"]):
        return None

    if not by_category:
        # Single overall average
        avg = round(df["Amount"].mean(), 2)
        total_txns = len(df)
        print(f"\n💳 AVERAGE SPENDING PER TRANSACTION")
        print(f"   Transactions analysed : {total_txns}")
        print(f"   Average amount        : ${avg:,.2f}\n")
        return avg

    else:
        # Per-category breakdown
        category_avg = (
            df.groupby("Category", as_index=False)["Amount"]
            .agg(Avg_Amount="mean", Num_Transactions="count")
            .sort_values("Avg_Amount", ascending=False)
            .reset_index(drop=True)
        )
        category_avg["Avg_Amount"] = category_avg["Avg_Amount"].round(2)

        print("\n💳 AVERAGE SPENDING PER TRANSACTION BY CATEGORY")
        print("-" * 50)
        print(f"  {'Category':<20} {'Avg Amount':>12}  {'# Txns':>7}")
        print("-" * 50)
        for _, row in category_avg.iterrows():
            print(f"  {row['Category']:<20} ${row['Avg_Amount']:>10,.2f}"
                  f"  {row['Num_Transactions']:>7}")
        print()

        return category_avg
````

````python
# ------------------------------------------------------------
# FUNCTION 4: Identify Unusual Transactions
# ------------------------------------------------------------

def find_unusual_transactions(df, z_threshold=2.0):
    """
    Flag transactions that are statistically unusual — either
    suspiciously large or negative (refunds / potential errors).

    Business Purpose:
        Unusual transactions can indicate fraud, billing errors, or
        one-off large expenses that distort the monthly budget.
        Flagging them lets the user review and categorise them
        appropriately before drawing conclusions from their data.

    How it works:
        Two detection methods are combined:

        1. Z-score method (statistical outliers):
           A Z-score measures how many standard deviations a value
           sits from the mean. A score above z_threshold (default 2.0)
           means the transaction is unusually large relative to the
           user's normal spending pattern.

           Z = (Amount - Mean) / Standard Deviation

        2. Negative amount detection:
           Negative values may be refunds, reversed charges, or data
           entry errors — all worth reviewing.

    Parameters:
        df          (pd.DataFrame): Cleaned transaction DataFrame.
        z_threshold (float)       : Z-score cutoff for "high" outliers
                                    (default 2.0 ≈ top ~2.5% of values).

    Returns:
        pd.DataFrame: Subset of df containing flagged rows, with an
                      extra "Flag" column explaining why each row was
                      flagged. Returns None if the input is invalid.
    """
    if not _validate_df(df, ["Amount", "Date", "Category", "Description"]):
        return None

    flagged_frames = []

    # -- Method 1: Statistical outliers (high Z-score) --------------
    mean   = df["Amount"].mean()
    std    = df["Amount"].std()

    if std == 0:
        # All amounts are identical — no variance to measure
        print("[INFO] All transaction amounts are identical; "
              "Z-score analysis skipped.")
    else:
        df_copy = df.copy()
        df_copy["_zscore"] = (df_copy["Amount"] - mean) / std
        high_outliers = df_copy[df_copy["_zscore"] > z_threshold].copy()
        high_outliers["Flag"] = (
            "High amount (Z=" + high_outliers["_zscore"].round(2).astype(str) + ")"
        )
        high_outliers = high_outliers.drop(columns=["_zscore"])
        flagged_frames.append(high_outliers)

    # -- Method 2: Negative amounts ---------------------------------
    negatives = df[df["Amount"] < 0].copy()
    negatives["Flag"] = "Negative amount (possible refund or error)"
    flagged_frames.append(negatives)

    # -- Combine & deduplicate (a row may trigger both methods) -----
    if not flagged_frames:
        print("[INFO] No unusual transactions detected.")
        return pd.DataFrame()

    unusual = (
        pd.concat(flagged_frames)
        .drop_duplicates(subset=df.columns.tolist())   # keep Flag intact
        .sort_values("Amount", ascending=False)
        .reset_index(drop=True)
    )

    if unusual.empty:
        print("\n✅ No unusual transactions detected.")
    else:
        print(f"\n⚠️  UNUSUAL TRANSACTIONS DETECTED ({len(unusual)} found)")
        print("-" * 70)
        for _, row in unusual.iterrows():
            print(f"  {str(row['Date'].date()):<12}  "
                  f"${row['Amount']:>10,.2f}  "
                  f"{row['Category']:<18}  "
                  f"{row['Flag']}")
        print()

    return unusual
````

````python
# AI-generated business insights
# (Your AI collaboration code goes here)

# ============================================================
# BLOCK 3: Professional Financial Report Generator
# Smart Finance Assistant – ISYS2001 Final Project
# ============================================================
# Business Purpose:
#   Combines the raw analysis from Block 2 with an AI-generated
#   narrative to produce a report a real financial planner might
#   hand to a client. The AI reads the spending data and returns
#   observations and savings recommendations in plain English.
#
# Depends on:
#   - clean_df        (Block 1: Data Loading & Cleaning)
#   - All functions   (Block 2: Spending Analysis)
#
# hands-on-ai reference:
#   Module : hands_on_ai.chat
#   Call   : get_response(prompt, system=...) → str
#   Docs   : https://michael-borck.github.io/hands-on-ai/
# ============================================================

import os
import pandas as pd
from datetime import datetime


# ------------------------------------------------------------
# STEP 1: Configure the hands-on-ai provider
# ------------------------------------------------------------
# These environment variables must be set before the first
# get_response() call. Setting them here keeps all config in
# one visible place at the top of the block.
#
# Per the project spec:
#   Server  : https://ollama.locollm.org  (/v1 is auto-appended)
#   Model   : gemma3:4b
#   API Key : Curtin2026ISYS20015002

os.environ['HANDS_ON_AI_SERVER']  = 'https://ollama.locollm.org'
os.environ['HANDS_ON_AI_MODEL']   = 'gemma3:4b'
os.environ['HANDS_ON_AI_API_KEY'] = 'Curtin2026ISYS20015002'

print("🔑 Hands-on-AI configured successfully!")

# ------------------------------------------------------------
# HELPER: Format a DataFrame as a plain-text table
# ------------------------------------------------------------

def _df_to_text(df, max_rows=10):
    """
    Convert a DataFrame to a compact plain-text table for embedding
    inside an LLM prompt.

    The model receives plain text — it cannot inspect Python objects —
    so we serialise just the rows and columns it needs to reason about.
    max_rows caps the output to keep the prompt concise and within the
    model's context window.

    Parameters:
        df       (pd.DataFrame): Table to format.
        max_rows (int)         : Maximum rows to include (default 10).

    Returns:
        str: Plain-text representation of the DataFrame.
    """
    return df.head(max_rows).to_string(index=False)

# ------------------------------------------------------------
# STEP 2: Collect all analysis results from Block 2
# ------------------------------------------------------------

def collect_analysis(df):
    """
    Run every Block 2 analysis function and bundle the results
    into a single dictionary. This gives the report generator
    one clean object to work with rather than scattered variables.

    Why a dictionary?
        Passing a single dict through the pipeline (collect →
        build_prompt → generate → assemble) is cleaner than
        managing five separate variables and makes it easy to
        add new analysis results later without changing the
        function signatures downstream.

    Parameters:
        df (pd.DataFrame): Cleaned transaction DataFrame from Block 1.

    Returns:
        dict: Keyed results from all four Block 2 functions, plus
              summary statistics for the report header.
              Returns None if any critical analysis step fails.
    """
    print("⏳ Running spending analysis...")

    # Call each Block 2 function and store its return value.
    # The functions also print their own summaries to the notebook;
    # here we focus on the returned DataFrames for the report.
    category_totals = spending_by_category(df)
    top_cats        = top_spending_categories(df, n=5)
    overall_avg     = average_spending_per_transaction(df)
    category_avg    = average_spending_per_transaction(df, by_category=True)
    unusual_txns    = find_unusual_transactions(df, z_threshold=2.0)

    if category_totals is None or top_cats is None:
        print("[ERROR] Analysis failed — cannot generate report.")
        return None

    # Compute a few extra summary stats used in the report header
    # and embedded in the LLM prompt
    summary = {
        "date_from"      : df["Date"].min().strftime("%d %b %Y"),
        "date_to"        : df["Date"].max().strftime("%d %b %Y"),
        "total_txns"     : len(df),
        "total_spend"    : round(df["Amount"].sum(), 2),
        "overall_avg"    : overall_avg,
        "num_categories" : df["Category"].nunique(),
        "unusual_count"  : len(unusual_txns) if unusual_txns is not None else 0,
    }

    return {
        "summary"         : summary,
        "category_totals" : category_totals,
        "top_cats"        : top_cats,
        "category_avg"    : category_avg,
        "unusual_txns"    : unusual_txns,
    }


# ------------------------------------------------------------
# STEP 3: Build the LLM prompt from analysis data
# ------------------------------------------------------------

def build_prompt(results):
    """
    Serialise the analysis results into a structured prompt that
    instructs the LLM to act as a financial advisor and produce
    spending observations and savings recommendations.

    Why build the prompt in a separate function?
        Keeping prompt construction isolated means it can be tested
        independently (print it, inspect it, iterate on it) without
        making an API call every time. It also makes it easy to
        adjust the output format without touching the API logic.

    Prompt design decisions:
        - Data is embedded as plain text tables (the model cannot
          read Python objects or DataFrames directly).
        - The four-section structure (Overview, Patterns, Savings,
          Flags) maps directly to what a financial planner would
          produce for a client.
        - "Use ONLY the data provided" prevents the model from
          inventing figures that do not appear in the user's CSV.

    Parameters:
        results (dict): Output from collect_analysis().

    Returns:
        str: A complete, ready-to-send prompt string.
    """
    s  = results["summary"]
    ct = _df_to_text(results["category_totals"])
    tc = _df_to_text(results["top_cats"])
    ca = _df_to_text(results["category_avg"])

    # Build the unusual transactions section only if rows were flagged
    if results["unusual_txns"] is not None and not results["unusual_txns"].empty:
        unusual_section = (
            f"UNUSUAL / FLAGGED TRANSACTIONS ({s['unusual_count']} found):\n"
            + _df_to_text(
                results["unusual_txns"][
                    ["Date", "Amount", "Category", "Description", "Flag"]
                ]
            )
        )
    else:
        unusual_section = "UNUSUAL TRANSACTIONS: None detected."

    prompt = f"""
You are a professional personal finance advisor preparing a client summary report.
Use ONLY the data provided below — do not invent figures.

=== CLIENT SPENDING DATA ===
Period         : {s['date_from']} to {s['date_to']}
Transactions   : {s['total_txns']}
Total Spend    : ${s['total_spend']:,.2f}
Average Txn    : ${s['overall_avg']:,.2f}
Categories     : {s['num_categories']}

SPENDING BY CATEGORY (all categories, $ totals):
{ct}

TOP 5 SPENDING CATEGORIES (with % of total spend):
{tc}

AVERAGE TRANSACTION AMOUNT BY CATEGORY:
{ca}

{unusual_section}
=== END OF DATA ===

Using the data above, write a professional financial summary report with these
four clearly labelled sections:

1. SPENDING OVERVIEW
   Two to three sentences summarising the overall spending picture for the period.
   Mention total spend, number of transactions, and date range.

2. KEY SPENDING PATTERNS
   Three to five dot-point observations about where money is going.
   Reference specific categories and dollar amounts from the data.

3. SAVINGS OPPORTUNITIES
   Three to five concrete, actionable recommendations the client could act on
   this month. Each recommendation should name the category and suggest a
   realistic target reduction or behaviour change.

4. FLAGS FOR REVIEW
   Comment on any unusual or flagged transactions. If none were found, say so.
   Suggest what the client should verify or follow up on.

Write in a professional but approachable tone. Use dollar signs and percentages
where they add clarity. Do not repeat the raw data tables — synthesise them
into insight.
"""
    return prompt.strip()


# ------------------------------------------------------------
# STEP 4: Generate the AI narrative via hands-on-ai
# ------------------------------------------------------------

def generate_ai_narrative(prompt):
    """
    Send the structured prompt to the LLM using get_response()
    from hands_on_ai.chat and return the financial advisor
    narrative it produces.

    hands-on-ai API used:
        get_response(prompt, system=...) → str

        - prompt : the user-facing content (our data + instructions)
        - system : sets the model's persona for the whole response
                   (the financial advisor role)
        - Returns a plain string — no parsing needed

    The system prompt establishes the advisor persona so the model
    maintains a consistent, professional tone regardless of what
    the user's transaction data contains.

    Parameters:
        prompt (str): Ready-to-send prompt string from build_prompt().

    Returns:
        str: The LLM's narrative response, or a descriptive error
             message if the API call fails so the report can still
             be printed with a placeholder.
    """
    system_prompt = (
        "You are a certified personal finance advisor. "
        "You write clear, accurate, and actionable financial summaries "
        "based strictly on the data provided to you. "
        "You never invent figures, and you always recommend realistic, "
        "achievable changes rather than generic advice."
    )

    print("🤖 Generating AI financial narrative...")

    try:
        # get_response signature (hands-on-ai 0.4.0):
        # get_response(prompt, model=None, system="...", personality="friendly",
        #              stream=False, retries=2, return_usage=False)
        #
        # We pass only prompt and system here — the model is inherited
        # from the HANDS_ON_AI_MODEL environment variable set in Step 1.
        response = get_response(
            prompt,
            system=system_prompt,
        )
        return response

    except Exception as e:
        # Return the error as a string so the report still assembles
        # and the user can see what went wrong without a full crash
        return (
            f"[ERROR] Could not generate AI narrative: {e}\n"
            "Please check your server connection with: handsonai doctor"
        )


# ------------------------------------------------------------
# STEP 5: Assemble and print the full report
# ------------------------------------------------------------

def generate_financial_report(df):
    """
    Master function — orchestrates all steps to produce and
    display the complete professional financial report.

    Execution flow (hub-and-spoke):
        main block
            → collect_analysis()    [calls all Block 2 functions]
            → build_prompt()        [serialises results to text]
            → generate_ai_narrative() [calls hands-on-ai LLM]
            → assemble + print report
            → return report string  [for Gradio UI in Block 7]

    Parameters:
        df (pd.DataFrame): Cleaned transaction DataFrame from Block 1.

    Returns:
        str: The complete formatted report as a single string so it
             can be passed to a Gradio Textbox or Markdown component
             in Block 7 without any reformatting.
    """
    # -- Collect analysis -------------------------------------------
    results = collect_analysis(df)
    if results is None:
        return None

    s = results["summary"]

    # -- Get AI narrative -------------------------------------------
    prompt    = build_prompt(results)
    narrative = generate_ai_narrative(prompt)

    # -- Timestamp for the report header ----------------------------
    generated_at = datetime.now().strftime("%d %B %Y, %I:%M %p")

    # -- Assemble the report ----------------------------------------
    report = f"""
╔══════════════════════════════════════════════════════════════╗
║           SMART FINANCE ASSISTANT — SPENDING REPORT          ║
╠══════════════════════════════════════════════════════════════╣
║  Period    : {s['date_from']} → {s['date_to']:<32} ║
║  Generated : {generated_at:<48} ║
╚══════════════════════════════════════════════════════════════╝

━━━ QUICK STATS ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
  Total Spend    : ${s['total_spend']:>10,.2f}
  Transactions   : {s['total_txns']:>10}
  Average Txn    : ${s['overall_avg']:>10,.2f}
  Categories     : {s['num_categories']:>10}
  Flagged Items  : {s['unusual_count']:>10}
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

━━━ TOP 5 SPENDING CATEGORIES ━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
"""

    # Append the top-5 table with ASCII progress bars
    for _, row in results["top_cats"].iterrows():
        bar = "█" * int(row["Percentage"] / 5)
        report += (
            f"  {row['Category']:<20} ${row['Total_Amount']:>9,.2f}"
            f"  ({row['Percentage']}%) {bar}\n"
        )

    report += f"""
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━

━━━ AI FINANCIAL ADVISOR SUMMARY ━━━━━━━━━━━━━━━━━━━━━━━━━━━━

{narrative}

━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
⚠️  DISCLAIMER: This report is generated by an AI assistant for
   educational purposes only. It is not professional financial
   advice. Please consult a qualified financial advisor before
   making any significant financial decisions.
━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━━
"""

    print(report)
    return report


# ------------------------------------------------------------
# MAIN – Generate the report
# ------------------------------------------------------------
# Assumes clean_df is in scope from Block 1 and all Block 2
# functions are defined. The returned string is stored in
# financial_report so it can be fed directly into the
# Gradio UI in Block 7 without re-running the analysis.

financial_report = generate_financial_report(clean_df)
````
